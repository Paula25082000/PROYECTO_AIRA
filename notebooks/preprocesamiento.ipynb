```python
# Read and display the first 20 rows of the AIRA dataset
print(aira.head(20))
```

```python
# Clean the AIRA dataset to include only the specified columns
aira_clean = aira[['Measure code', 'AIRA_SIMPLE', 'COUNTRY_REGION']].copy()

print(f"Cleaned AIRA dataset shape: {aira_clean.shape}")
print(f"Number of rows: {aira_clean.shape[0]}")
print(f"Number of columns: {aira_clean.shape[1]}")
print(f"\nFirst few rows:")
print(aira_clean.head())
```

```python
# Filter AIRA dataset for Measure code AIRA_1 and display results by country
aira_1_table = aira_clean[aira_clean['Measure code'] == 'AIRA_1'].copy()

# Pivot the table to show countries and their responses
aira_1_pivot = aira_1_table.pivot_table(
    values='AIRA_SIMPLE', 
    index='COUNTRY_REGION', 
    aggfunc='first'
).reset_index()

# Rename columns for clarity
aira_1_pivot.columns = ['Country/Region', 'Response']

# Sort by country name
aira_1_pivot = aira_1_pivot.sort_values('Country/Region').reset_index(drop=True)

print(f"AIRA_1: National AI strategy, policy, action plan or equivalent for AI in the health sector")
print(f"\nTotal countries/regions: {len(aira_1_pivot)}")
print("\n")
print(aira_1_pivot)
```
